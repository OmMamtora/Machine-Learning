```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
titanic = pd.read_csv('titanic.csv')
titanic.head()
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

## 1) Scatter Plot (Numerical - Numerical)

### Bia variant

```python
sns.scatterplot(x=tips['total_bill'], y = tips['tip'])
```

### Multivariant

```python
sns.scatterplot(x = tips['total_bill'], y = tips['tip'], hue=tips['sex'], style = tips['smoker'], size = tips['size'])
```

## 2) Box Plot (Numerical - Categorical)

```python
sns.barplot(x = titanic['Pclass'], y = titanic['Age'])
```

```python
sns.barplot(x = titanic['Pclass'], y = titanic['Fare'])
```

```python
sns.barplot(x = titanic['Pclass'], y = titanic['Age'], hue = titanic['Sex'])
```

```python
sns.barplot(x = titanic['Pclass'], y = titanic['Fare'], hue = titanic['Embarked'])
```

### 3) Box Plot (Numerical - Categorical)

```python
sns.boxplot(x = titanic['Sex'], y = titanic['Age'])
```

```python
sns.boxplot(x = titanic['Sex'], y = titanic['Age'], hue = titanic['Survived'])
```

### 4) Distplot (Numerical - Categorical)

```python
sns.distplot(x = titanic['Survived'])
```

```python
sns.distplot(x = titanic[titanic['Survived'] == 0]['Age'])
sns.distplot(x = titanic[titanic['Survived'] == 1]['Age'])
```

```python
sns.distplot(x = titanic[titanic['Survived'] == 0]['Age'], hist = False)
sns.distplot(x = titanic[titanic['Survived'] == 1]['Age'], hist = False)
plt.legend(titanic['Survived'])
```

### 5) Heatmap (Categorical - Categorical)

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']), cmap = 'coolwarm')
```

### 6) Clustermap (Categorical - Categorical)

```python
pd.crosstab(titanic['Parch'], titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']),)
```

### 7) Pairplot

```python
sns.pairplot(iris, hue = 'species')
```

### 8) Lineplot (Numerical - Numerical)

```python
sns.lineplot(x = flights['year'] ,y = flights['passengers'])
```

```python
flights.pivot_table(values = 'passengers', index = 'month', columns = 'year')
```

```python
sns.heatmap(flights.pivot_table(values = 'passengers', index = 'month', columns = 'year'))
```

```python
sns.clustermap(flights.pivot_table(values = 'passengers', index = 'month', columns = 'year'))
```
