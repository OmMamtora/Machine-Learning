# 4. OneHotEncoding with top Categories

### Here, i am going to do OHE on brand and club a car categories whoes lessthan 100

```python
df['brand'].value_counts()
```

```python
counts = df['brand'].value_counts()
```

```python
df['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts <= threshold].index
```

```python
repl.shape
```

```python
pd.get_dummies(df['brand'].replace(repl, 'uncommon'), dtype = int)
```

```python
pd.get_dummies(df['brand'].replace(repl, 'uncommon'), dtype = int).sample(10)
```
