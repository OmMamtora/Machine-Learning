```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('wine_data.csv',header=None, usecols = [0,1,2])
df.columns=['Class label','Alcohol','Malic acid']
```

```python
df.head()
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic acid'])
```

```python
sns.scatterplot(x = df['Alcohol'], y = df['Malic acid'], hue=df['Class label'],palette = "tab10")
```

```python
color_dict = {1 : 'red', 3 : 'green' , 2 : 'blue'}
sns.scatterplot(x = df['Alcohol'], y = df['Malic acid'], hue = df['Class label'], palette = color_dict)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split


X_train, X_test, Y_train, Y_test = train_test_split(df.drop('Class label', axis = 1),
                                                           df['Class label'],
                                                           test_size=0.3,
                                                           random_state=0)

X_train.shape, X_test.shape
```

## MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler

scale = MinMaxScaler()

scale.fit(X_train)

X_train_scaled = scale.transform(X_train)
X_test_scaled = scale.transform(X_test)
```

```python
X_train
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_train_scaled
```

```python
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns)
X_test_scaled
```

```python
np.round(X_train.describe(),1)
```

```python
np.round(X_train_scaled.describe(),1)
```

```python
fig, axes = plt.subplots(ncols = 2, figsize=(12,5))

axes[0].set_title("Before Scaling")
sns.scatterplot(x = X_train['Alcohol'], y = X_train['Malic acid'], c = Y_train, ax=axes[0])


axes[1].set_title("After Scaling")
sns.scatterplot(x = X_train_scaled['Alcohol'], y = X_train_scaled['Malic acid'], c = Y_train, ax=axes[1])


plt.show()
```

```python
fig , axes = plt.subplots(ncols = 2, figsize = (12,5))

axes[0].set_title("Before Scaling")
sns.kdeplot(X_train['Alcohol'], ax= axes[0])
sns.kdeplot(X_train['Malic acid'], ax = axes[0])


axes[1].set_title("After Scaling")
sns.kdeplot(X_train_scaled['Alcohol'], ax = axes[1])
sns.kdeplot(X_train_scaled['Malic acid'], ax = axes[1])

plt.show()
```
